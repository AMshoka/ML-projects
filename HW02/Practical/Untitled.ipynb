```python
def normal_params(df):
    covMatrix = pd.DataFrame.cov(df)
    sigma=covMatrix.values
    sigma=np.matrix(sigma)
    mu=df.mean()
    mu=np.array(mu)

    return sigma,mu
```

```python
def Accuracy(y_true,y_pred):
    true=np.size(np.where(y_pred==y_true)) 
    acc=true/len(y_true)*100
    return acc
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
توابع زیر با گرفتن دو بردار که یکی همان 
target 
و دیگری همان  
prediction
است، معیارهای خواسته شده را محاسبه می‌کنند.
</font></div>

```python
def cal_pred_detail(y_true,y_pred):
    tp=0
    tn=0
    fp=0
    fn=0
    for i in range (len(y_true)):
        if (y_pred[i]==y_true[i] and y_pred[i]==0):
            tn=tn+1
        elif (y_pred[i]==y_true[i] and y_pred[i]==1):
            tp=tp+1
        elif (y_pred[i]!=y_true[i] and y_pred[i]==0):
            fn=fn+1
        else:
            fp=fp+1
    return tp,tn,fp,fn
        
    
```

```python
def precision(y_true, y_pred):
    tp,tn,fp,fn=cal_pred_detail(y_true,y_pred)
    p=tp/(tp+fp)*100
    return p
```

```python
def recall(y_true, y_pred):
    tp,tn,fp,fn=cal_pred_detail(y_true,y_pred)
    r=tp/(tp+fn)*100
    return r
```

```python
def f1score(y_true, y_pred):
    tp,tn,fp,fn=cal_pred_detail(y_true,y_pred)
    p=tp/(tp+fp)
    r=tp/(tp+fn)
    f=2*p*r/(p+r)*100
    
    return f
```

```python
class LR() :
    def __init__( self, lr, iterations,lam ) :        
        self.lr = lr
        self.lam=lam
        self.iterations = iterations
              
    def f(self):
        xw=self.X.dot( self.W )
        return xw
    
    def l2(self):
        lr2=2*self.W
        return lr2
        
        
    def sigm(self):
        si = 1 / ( 1 + np.exp( - ( self.f() + self.b ) ) ) 
        return si
        
      
    def dj( self ) : 
        
        si=self.sigm()       
        tmp = ( si - self.Y.T )        
        tmp = np.reshape( tmp, self.m )        
        dW = (np.dot( self.X.T, tmp ) / self.m)+(self.lam*self.l2())
        db = np.sum( tmp ) / self.m 
            
        self.W = self.W - self.lr * dW     
        self.b = self.b - self.lr * db
          
        return self
    
    def fit( self, X, Y ) :        
        print(X.shape)
        print(Y.shape)
        self.m, self.n = X.shape                
        self.W = np.zeros( self.n )        
        self.b = 0        
        self.X = X        
        self.Y = Y
                  
        for i in range( self.iterations ) :            
            self.dj()            
        return self,self.W
      
 
      
    def classify( self, X ) :    
        Z = 1 / ( 1 + np.exp( - ( X.dot( self.W ) + self.b ) ) )        
        Y = np.where( Z > 0.5, 1, 0 )        
        return Y
  
```

```python
def euclidean(a,b):
    return np.linalg.norm(a-b)
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
تابع زیر یک بردار و یک دیتافریم می‌گیرد و فاصله‌ی آن بردار با تمام بردار‌های موجود در دیتافریم را برمی‌گرداند
</font></div>

```python
def distance(x,m,X_train,Y_train,k):
    euclidean_distances = np.zeros(m)
    for i in range(m) :
        d = euclidean( x, X_train[i] )
        euclidean_distances[i] = d
    inds = euclidean_distances.argsort()
    Y_train_sorted = Y_train[inds]

    return Y_train_sorted[:K]
```

```python
def knn_decide(neighbors):
    return mode(neighbors)[0][0]  
```

```python
class KNN() : 
      
    def __init__( self, K ) :
        self.K = K
          
    def fit( self, X_train, Y_train ) :
        self.X_train = X_train
        self.Y_train = Y_train
        self.m, self.n = X_train.shape
        
    def distance( self, x ) :
        euclidean_distances = np.zeros( self.m )
        for i in range( self.m ) :
            d = self.euclidean( x, self.X_train[i] )
            euclidean_distances[i] = d
        inds = euclidean_distances.argsort()
        Y_train_sorted = self.Y_train[inds]
        return Y_train_sorted[:self.K]
    
    def euclidean( self, x, x_train ):
          
        return np.linalg.norm(x-x_train)
    
    def knn_decide(self, neighbors):
         return mode( neighbors )[0][0]  
        
    def nearest( self, X_test ) :
        self.X_test = X_test
        self.m_test, self.n = X_test.shape
        Y_nearest = np.zeros( self.m_test )
        for i in range( self.m_test ) :
            x = self.X_test[i]
            neighbors = np.zeros( self.K )
            neighbors = self.distance( x )
            Y_nearest[i] = self.knn_decide(neighbors)  
              
        return Y_nearest
```

```python
def preprocessing(train):
    X = train.iloc[:,:-1].values
    Y = train.iloc[:,-1:].values
    X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size = 0.2, random_state = 0 )
    return X_train, Y_train, X_val, Y_val
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
تابع زیر داده‌ی آموزش و هایپرپارامتر‌های موردنیاز را (که به صورت دیکشنری است و یکی از کلید‌های آن layers بوده که مقدار آن  آرایه‌ای است شامل پارامتر‌های مربوط به هر لایه)را می‌گیرد
    و پس از پیاده‌سازی مدل و آموزش آن, مدل ترین‌شده را بازمی‌گرداند
</font></div>

```python
def train_model(X_train, y_train, hparams):
    model = Sequential()
    model.add(Dense(hparams[0], input_dim=13, kernel_initializer='random_uniform', activation='relu'))
    model.add(Dense(hparams[1], input_dim=13, kernel_initializer='random_uniform', activation='relu'))
    model.add(Dense(hparams[2], input_dim=13, kernel_initializer='random_uniform', activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.fit(X_train,y_train,epochs=100,batch_size=10)
    return model
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
تابع زیر در ورودی, مدل ترین‌شده و داده‌ی تست(یا اعتبارسنجی) را می‌گیرد و در خروجی, دقت مدل را برمی‌گرداند. از خروجی این تابع می‌توانید برای بررسی عملکرد مدل  استفاده کنید.</font></div>

```python
def test_model(model, X_test, y_test):
    scores = model.evaluate(X_test,y_test)
    return scores[1]
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
حال با استفاده از توابع بالا کل فرآیند را در ذیل توسعه دهید.
</font></div>

```python
hparams=[[15,7,1],[50,25,1],[100,50,1]]
best_hparam = []
val_accs = [] #compare results based on validation set
X_train, y_train, X_val, y_val=preprocessing(train)
for param in hparams:
    model=train_model(X_train,y_train,param)
    print('***************Validation*******************')
    scores = model.evaluate(X_val,y_val)
    print('********************************************')
    val_accs.append(scores)
    #test_acc = test_model(model, x_test, y_test)
    #print("Model Test Accuracy: {}".format(test_acc))

for i in range (len(val_accs)):
    if (val_accs[i][1]==np.max(val_accs)):
        best_hparam=hparams[i]
        
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
<b style="font-size: 20px">ارزیابی (دستیار آموزشی)</b>
<br>
    کد شما در این بخش ارزیابی می‌شود. در این قسمت فارغ از دانستن داده‌ها, پس از طی کردن مراحل ترین و تست, مدل با بهترین نتایج را به تابع زیر پاس دهید و دقت آنرا چاپ کنید. 
</font></div>

```python
def predict_test_TA(X,Y,best_hparam,x_test,y_test):
    model=train_model(X,Y,best_hparam)
    test_acc = test_model(model, x_test, y_test)
    return test_acc
```
